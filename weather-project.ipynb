# 16. line chart showing the temprature over time .

```python
# convert date/time column into date/time formate
data['Date/Time'] = pd.to_datetime(data['Date/Time'])

#plotting the temprature over time 
plt.figure(figsize=(10,6))
plt.plot(data['Date/Time'],data['Temp_C'],color='blue')
plt.title('Temperature over time')
plt.xlabel('Date/Time')
plt.ylabel('Temperature (c)')
plt.grid(True)
plt.show()
```

# 17.line chart showing the variation of humidity over time

```python
# Assuming 'Date/Time' is already converted to datetime in a previous cell
# If not, uncomment the next line:
# data['Date/Time'] = pd.to_datetime(data['Date/Time'])

# Plotting the humidity over time
plt.figure(figsize=(10, 6))
plt.plot(data['Date/Time'], data['Rel Hum_%'], color='green')
plt.title('Humidity over time')
plt.xlabel('Date/Time')
plt.ylabel('Relative Humidity (%)')
plt.grid(True)
plt.show()
```

# 18. bar chart showing the distribution of weather condition

```python
# Bar chart showing the distribution of weather conditions
weather_counts = data['weather_condition'].value_counts()
plt.figure(figsize=(10, 6))
weather_counts.plot(kind='bar', color='skyblue')
plt.title('Distribution of Weather Conditions')
plt.xlabel('Weather Condition')
plt.ylabel('Frequency')
plt.xticks(rotation=90)
plt.show()
```

# 19. box plot showing the distribution of tempreture for each weather condition:

```python
# Box plot showing the distribution of temperature for each weather condition
plt.figure(figsize=(10, 6))
sns.boxplot(x='weather_condition', y='Temp_C', data=data)
plt.title('Temperature Distribution by Weather Condition')
plt.xticks(rotation=90)
plt.show()
```

# 20.bar plot of pressure 

```python
# Bar plot of average pressure by weather condition
pressure_by_weather = data.groupby('weather_condition')['Press_kPa'].mean()
plt.figure(figsize=(10, 6))
pressure_by_weather.plot(kind='bar', color='steelblue')
plt.title('Average Pressure by Weather Condition')
plt.xlabel('Weather Condition')
plt.ylabel('Average Pressure (kPa)')
plt.xticks(rotation=90)
plt.show()
```

# 21. box plot of visibility 

```python
# Box plot of visibility by weather condition
plt.figure(figsize=(10, 6))
sns.boxplot(x='weather_condition', y='Visibility_km', data=data)
plt.title('Visibility Distribution by Weather Condition')
plt.xlabel('Weather Condition')
plt.ylabel('Visibility (km)')
plt.xticks(rotation=90)
plt.show()
```

# 22. calculate the value counts for each weather condition using pi cahrt

```python
# Calculate the value counts for each weather condition using pie chart
weather_counts = data["weather_condition"].value_counts()
plt.figure(figsize=(10, 6))
plt.pie(weather_counts, labels=weather_counts.index, autopct='%1.1f%%', startangle=90 , pctdistance=0.85)
plt.title('Distribution of Weather Conditions (Pie Chart)')
plt.show()
```

# 23.scatter plot of temprature vs humidity

```python
# Scatter plot of temperature vs humidity colored by weather condition
plt.figure(figsize=(10, 6))
for weather in data['weather_condition'].unique():
    mask = data['weather_condition'] == weather
    plt.scatter(data[mask]['Temp_C'], data[mask]['Rel Hum_%'], label=weather, alpha=0.6)
plt.title('Temperature vs Humidity by Weather Condition')
plt.xlabel('Temperature (°C)')
plt.ylabel('Relative Humidity (%)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

# 24.histogram of wind speed

```python
# Histogram of wind speed
plt.figure(figsize=(10, 6))
plt.hist(data['Wind Speed_km/h'], bins=20, color='coral', edgecolor='black')
plt.title('Distribution of Wind Speed')
plt.xlabel('Wind Speed (km/h)')
plt.ylabel('Frequency')
plt.grid(True, alpha=0.3)
plt.show()
```
